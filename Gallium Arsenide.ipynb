Now with we learnt the basics of the Bandstructure in [[Bandstructure]], we can can move to the common semi-conductors like Gallium Arsenide.
- All the common semiconductors (like gallium arsenide) belong to the diamond structure which has a unit cell consisting of two atoms, a cation (like gallium) and an anion (like arsenic).
- For elemental semiconductors like silicon, both cationic and anionic sites are occupied by the same atom

- For each atom we need to include at least four valence orbitals like $3s,3p_{x},3p_{y},3p_{z}$ for silicon, and also the next $4s$ orbital, to give rise to the $sp^3s^\star$ model. So we have 5 orbitals per atom, leading to 10 basis orbitals per unit cell. 


- **In the FCC, we have Gallium occupying the FCC sites and Arsenic occupying the tetrahedral voids.** 

The arsenic atoms occupy the sites of a different FCC lattice offset from the previous one by a **quarter of the distance along the body diagonal**– that is, the coordinates of this lattice can be obtained by adding $(\hat{x}+\hat{y}+\hat{z})a/4$ to those of the first one.

- So if a gallium atom is located at the origin $(0,0,0)a/4$ then there will be arsenic atom located at the $(\hat{x}+\hat{y}+\hat{z})a/4$, which will be one of its nearest neighbors. Actually it will also have **three more arsenic atoms** as nearest neighbors.

Now we can see that we have 4 Ga atoms on the center of squares (face), whose coordinates can be written as $(\hat{x}+\hat{y})a/2, ~(-\hat{x}+\hat{y})a/2,~(-\hat{x}-\hat{y})a/2,~(\hat{x}-\hat{y})a/2$

- The coordinates of the corresponding arsenic atoms are obtained by adding, $(\hat{x}+\hat{y}+\hat{z})a/4$. 

$$(3\hat{x}+3\hat{y}+\hat{z})a/4~~~~(3\hat{x}-\hat{y}+\hat{z})a/4~~~~(-\hat{x}+3\hat{y}+\hat{z})a/4~~~~(-\hat{x}-\hat{y}+\hat{z})a/4~~~~$$

- Every atom in a diamond lattice has four nearest neighbors of the opposite type (cation or anion) arranged in a tetrahedron.

The Hamiltonian $H$ is expressed in a **2×2 matrix**, because the tight-binding basis includes:

- $|s_a\rangle$: the s-orbital on **anion** (As)
    
- $|s_c\rangle$: the s-orbital on **cation** (Ga)
$$H_{nm}=\begin{array}{c|ccc} 
 & \ket{s_{a}} & \ket{s_{c}}  \\
\hline \ket{s_{a}}  & E_{sa} & E_{ss} \\
\ket{s_{c}}  & E_{ss} & E_{sc} \\  
\end{array}$$
Where 
- $E_{sa}$​: on-site energy of s orbital on anion As
    
- $E_{sc}$​: on-site energy of s orbital on cation Ga
    
- $E_{ss}$​: overlap (or hopping) between s orbitals on Ga and As

These hopping integrals contribute to the **band dispersion** and are essential to capture the band structure near the valence and conduction band edges.

Now we need to imagine this:

The anion in unit cell $n$ overlaps with the cation in unit cell $n$ plus the cations in three other unit cells $m$ for which
$$\vec{d}_{m}-\vec{d}_{n}=(-\hat{x}-\hat{y})a/2~~~~~~(-\hat{y}-\hat{z})a/2~~~~~~(-\hat{z}-\hat{x})a/2$$

Each of these contributes a $[Hnm$] of the form:
$$H_{nm}=\begin{array}{c|ccc} 
 & \ket{m,s_{a}} & \ket{m,s_{c}}  \\
\hline \ket{n,s_{a}}  & 0 & E_{ss} \\
\ket{n,s_{c}}  & 0 & 0 \\  
\end{array}$$

Similarly the cation in the unit cell $n$ overlaps with the anion in the unit cell $n$ plus the anions in three other unit cells $m$ for which $$\vec{d}_{m}-\vec{d}_{n}=(\hat{x}+\hat{y})a/2~~~~~~(\hat{y}+\hat{z})a/2~~~~~~(\hat{z}+\hat{x})a/2$$

Each of these contributes a $[Hnm$] of the form:
$$H_{nm}=\begin{array}{c|ccc} 
 & \ket{m,s_{a}} & \ket{m,s_{c}}  \\
\hline \ket{n,s_{a}}  & 0 & 0 \\
\ket{n,s_{c}}  & E_{ss} & 0 \\  
\end{array}$$

***

Adding up these contributions we obtain:

$$[h(\vec{k})]=\begin{array}{c|cc} 
 & \ket{s_{a}} & \ket{s_{c}} \\
\hline \ket{s_{a}} & E_{sa} & 4E_{ss}g_{0} \\
\ket{s_{c}} & 4E_{ss}g_{0}^* & E_{sc}    
\end{array}$$

where $$g_{0}\equiv \frac{1}{4}(1+e^{-i\vec{k}\cdot \vec{d}_{1}}+e^{-i\vec{k}\cdot \vec{d}_{2}}+e^{-i\vec{k}\cdot \vec{d}_{3}})$$

with $$\vec{d}_{1}\equiv (\hat{y}+\hat{z})a/2~~~~~~ \vec{d}_{2}\equiv (\hat{z}+\hat{x})a/2~~~~~~\vec{d}_{3}\equiv (\hat{x}+\hat{y})a/2$$



$$\begin{array}{c|cccccccccc}
     & |s_a\rangle & |s_c\rangle & |X_a\rangle & |Y_a\rangle & |Z_a\rangle & |X_c\rangle & |Y_c\rangle & |Z_c\rangle & |s_a^*\rangle & |s_c^*\rangle \\
\hline
|s_a\rangle & E_{sa} & 4E_{ss}g_0 & 0 & 0 & 0 & 4E_{sapc}g_1 & 4E_{sapc}g_2 & 4E_{sapc}g_3 & 0 & 0 \\
|s_c\rangle & 4E_{ss}g_0^* & E_{sc} & 4E_{pasc}g_1^* & 4E_{pasc}g_2^* & 4E_{pasc}g_3^* & 0 & 0 & 0 & 0 & 0 \\
|X_a\rangle & 0 & 4E_{pasc}g_1 & E_{pa} & 0 & 0 & 4E_{xx}g_1 & 4E_{xy}g_3 & 4E_{xy}g_2 & 0 & 4E_{pas^*c}g_1 \\
|Y_a\rangle & 0 & 4E_{pasc}g_2 & 0 & E_{pa} & 0 & 4E_{xy}g_3 & 4E_{xx}g_2 & 4E_{xy}g_1 & 0 & 4E_{pas^*c}g_2 \\
|Z_a\rangle & 0 & 4E_{pasc}g_3 & 0 & 0 & E_{pa} & 4E_{xy}g_2 & 4E_{xy}g_1 & 4E_{xx}g_3 & 0 & 4E_{pas^*c}g_3 \\
|X_c\rangle & 4E_{sapc}g_1^* & 0 & 4E_{xx}g_1^* & 4E_{xy}g_3^* & 4E_{xy}g_2^* & E_{pc} & 0 & 0 & 4E_{s^*apc}g_1^* & 0 \\
|Y_c\rangle & 4E_{sapc}g_2^* & 0 & 4E_{xy}g_3^* & 4E_{xx}g_2^* & 4E_{xy}g_1^* & 0 & E_{pc} & 0 & 4E_{s^*apc}g_2^* & 0 \\
|Z_c\rangle & 4E_{sapc}g_3^* & 0 & 4E_{xy}g_2^* & 4E_{xy}g_1^* & 4E_{xx}g_3^* & 0 & 0 & E_{pc} & 4E_{s^*apc}g_3^* & 0 \\
|s_a^*\rangle & 0 & 0 & 0 & 0 & 0 & 4E_{s^*apc}g_1 & 4E_{s^*apc}g_2 & 4E_{s^*apc}g_3 & E_{s^*a} & 0 \\
|s_c^*\rangle & 0 & 0 & 4E_{pas^*c}g_1^* & 4E_{pas^*c}g_2^* & 4E_{pas^*c}g_3^* & 0 & 0 & 0 & 0 & E_{s^*c}
\end{array}
\quad (5.3.3)$$



````python
import numpy as np
from ase import Atoms
from ase.build import make_supercell
from ase.io import write
from io import BytesIO
import py3Dmol

# Define lattice constant
a = 5.65

# Base unit cell
ga_frac = [(0,0,0), (0.5,0,0.5), (0,0.5,0.5), (0.5,0.5,0)]
as_frac = [(x+0.25, y+0.25, z+0.25) for x,y,z in ga_frac]
ga_cart = [np.dot(f, np.eye(3)*a) for f in ga_frac]
as_cart = [np.dot(f, np.eye(3)*a) for f in as_frac]
symbols = ['Ga']*4 + ['As']*4
positions = ga_cart + as_cart
gaas = Atoms(symbols=symbols, positions=positions, cell=[a,a,a], pbc=True)

# Supercell
supercell = make_supercell(gaas, np.eye(3)*2)

# Convert to CIF string
cif_io = BytesIO()
write(cif_io, supercell, format='cif')
cif_str = cif_io.getvalue().decode()

# Add box atoms (8 corner dummy atoms, element H)
box_atoms = [
    [0,0,0], [a,0,0], [a,a,0], [0,a,0],
    [0,0,a], [a,0,a], [a,a,a], [0,a,a]
]
xyz_box = [f"{len(box_atoms)}\nBox"]
for pos in box_atoms:
    xyz_box.append(f"H {pos[0]:.3f} {pos[1]:.3f} {pos[2]:.3f}")
xyz_str = "\n".join(xyz_box)

# Set up py3Dmol view
view = py3Dmol.view(width=800, height=800)
view.addModel(cif_str, 'cif')

# Atom styles
view.setStyle({'elem': 'Ga'}, {
    'sphere': {'scale': 0.25, 'color': 'orange'},
    'stick': {'radius': 0.12, 'color': 'orange'}
})
view.setStyle({'elem': 'As'}, {
    'sphere': {'scale': 0.25, 'color': 'purple'},
    'stick': {'radius': 0.12, 'color': 'purple'}
})

# Add bounding box using dummy atoms
view.addModel(xyz_str, "xyz")
view.setStyle({'model': 1}, {'stick': {'radius': 0.05, 'color': 'gray'}})

# Final display
view.setBackgroundColor("white")
view.zoomTo()
view.show()
````

# First Band Structure - s orbital only

We’ve constructed the **simplest tight-binding model** for GaAs:  
A **two-band model**, with only **s-orbitals** from each atom:

- Gallium (Ga): cation → $s_{\text{Ga}}$​
    
- Arsenic (As): anion → $s_{\text{As}}$​
    

This gives a **2×2 Hamiltonian** at each $\vec{k}$-point in the Brillouin zone.

## 🧠 **1. GaAs Structure Basics**

- GaAs has a **zinc blende structure**, which is like the FCC lattice with a two-atom basis:
    
    - **Ga at (0, 0, 0)**
        
    - **As at (¼, ¼, ¼)** (in units of the lattice constant aaa)
        

Because there are **two atoms per unit cell**, the tight-binding Hamiltonian will always be **at least 2×2** when including only s-orbitals: one basis function per atom.
## 🧮 **2. The Hamiltonian**

For any wavevector $\vec{k}$, the Hamiltonian is:
$$H(\vec{k})=\begin{pmatrix}
E_{a} & 4E_{sa}\cdot g_{0}(\vec{k}) \\
4E_{ssa}\cdot g^*_{0}(\vec{k}) & E_{c}
\end{pmatrix}$$

- $E_a$​: on-site energy of **As** (anion, deeper potential → more negative)
    
- $E_c$​: on-site energy of **Ga**
    
- $E_{sa}$​: **hopping integral** between Ga and As sss-orbitals (negative in sign)
    
- g0(k⃗)g_0(\vec{k})g0​(k): structure factor encoding the phase factors from orbital overlaps
    

The **factor of 4** is due to 4 nearest-neighbor bonds between As and Ga in the zinc blende lattice.

## 📐 **3. Structure Factor $g_0(\vec{k})$**

You considered only **3 of the 4** displacement vectors d⃗i\vec{d}_idi​ between Ga and As atoms.

Why?

Because:

- The 4th one is at $\vec{0}$, meaning the atom within the same unit cell → gives $\exp(0) = 1$
    
- So the sum becomes:
    

$$g_0(\vec{k}) = 1 + \sum_{i=1}^{3} \exp(-i \vec{k} \cdot \vec{d}_i)$$
This gives a **complex number**, representing the combined phase overlap of Ga-As bonds at different directions.

## 🔬 **4. Band Energies**

From $H(\vec{k})$, compute eigenvalues:

$$E_{\pm}(\vec{k})=\frac{{E_{a}+E_{c}}}{2}\pm \sqrt{ \left( \frac{{E_{a}-E_{c}}}{2} \right)^2+\lvert 4E_{sa}g_{0}(\vec{k}) \rvert^2  }$$

- $E_{-}$​: **valence band** (mostly As $s$-character)
    
- $E_+$​: **conduction band** (mostly Ga $s$-character)
    

This is the **two-band s-orbital approximation**, with direct coupling.

## 🔁 **5. $\vec{k}$-Path and Plot**

You're plotting along the **standard high-symmetry path in the Brillouin zone**:
$$\Gamma\to X\to L\to\Gamma$$
Where:

- $\Gamma = (0,0,0)$
    
- $X = (1/2, 0, 0)$
    
- $L=(½,½,½)$

All points are in **units of $2\pi / a$** → you convert to real $\vec{k}$-space via:

$$\vec{k} = \left(\text{fractional coords}\right) \cdot \frac{2\pi}{a}$$

You calculate the **distance along the $\vec{k}$-path** to use for $x$-axis in the band structure.


#### 📐 ** What are High-Symmetry Points?**

These are special points in the BZ where:

- The crystal's symmetry operations (rotations, mirrors, etc.) leave the point invariant.
    
- The Hamiltonian has maximal degeneracies or special properties.
    
- They're used as standard reference points for comparing materials.
    

In FCC (which is the reciprocal lattice of GaAs’ real-space lattice), common **high-symmetry points** are:

| Symbol | $\vec{k}$-coordinates (in units of $2\pi/a$) | Description                         |
| ------ | -------------------------------------------- | ----------------------------------- |
| $Γ$    | (0, 0, 0)                                    | Center of the BZ                    |
| $X$    | (½, 0, 0)                                    | Face center of the cube             |
| $L$    | (½, ½, ½)                                    | Corner of the BZ (body diagonal)    |
| $K$    | (¾, ¾, 0)                                    | Edge of the hexagon (not used here) |
| $W$    | (1, ½, 0)                                    | Edge center                         |
| $U$    | (1, ¼, ¼)                                    | Inside edge center                  |

Each point lies at a region of **special geometric or symmetry property**.
#### Why plot this way?

You don't plot the entire 3D Brillouin zone — instead, you choose a **1D path** through it that passes through these high-symmetry points:

### Standard path for FCC (GaAs):

$$\Gamma \rightarrow X \rightarrow L \rightarrow \Gamma$$

This path:

- Captures **key symmetry directions** in the crystal
    
- Shows **band extrema** and possible band gap locations
    
- Matches convention (used in all research & materials databases)
    

🧭 This path gives you a **clear and interpretable projection of the 3D band structure**.

- $Γ→X$: along a **Cartesian axis**
    
- $X→L$: diagonal across a **face**
    
- $L→Γ$: back along a **body diagonal**
    

This way you **sample distinct directions** in the crystal to see how bands evolve.

#### **. Model Assumptions**

This model is a **minimal toy model**, but it captures:

- Indirect coupling between atoms via hopping
    
- Band splitting due to hybridization
    
- Direct band gap at $\Gamma$ (true for GaAs)

## 📉 **7. Interpreting the Output Plot**

- Two smooth bands:
    
    - **Blue** (valence) lies below red (conduction)
        
    - **Band gap** visible at $Γ$
        
- This is a **direct-gap semiconductor**, as GaAs is in reality.
    

The bands follow the general shape, but are too simplified for real quantitative prediction — for that, you need sp³ or sp³d⁵ tight-binding models.

## 📐 **4. Displacement Vectors for Hopping**

````python
d1 = (a / 4) * np.array([+1, +1, +1])
d2 = (a / 4) * np.array([+1, +1, -1])
d3 = (a / 4) * np.array([+1, -1, +1])
````

These are **3 of the 4 nearest-neighbor Ga–As displacement vectors**, from the As at (¼,¼,¼) to 3 surrounding Ga atoms.

- The 4th one is (−1,−1,−1), which is **within the same unit cell** and contributes `1` (so added manually in `g0`).

## 🔁 **5. Band Structure Calculation**

```python
E_plus = []
E_minus = []

for k in k_points:
    g0 = 1 \
        + np.exp(-1j * np.dot(k, d1)) \
        + np.exp(-1j * np.dot(k, d2)) \
        + np.exp(-1j * np.dot(k, d3))
```

 You calculate the **structure factor $g_0(\vec{k})$** using 3 displacement vectors.
    
- The `1` in front represents the **4th neighbor** at same cell (trivial phase: $e^{0} = 1$).

```python
    H_k = np.array([
        [E_a, 4 * E_sa * g0],
        [4 * E_sa * np.conj(g0), E_c]
    ])

```

You build the **2×2 Hamiltonian matrix**, with:

- Diagonal = on-site terms
    
- Off-diagonal = hopping (scaled by 4 for 4 neighbors)

```python
    eigvals = np.linalg.eigvalsh(H_k)
    E_minus.append(eigvals[0])
    E_plus.append(eigvals[1])

```

- We get **eigenvalues** $E_-(k)$ and $E_+(k)$ → band energies.


````python
# Band structure computation
E_plus = []
E_minus = []

for k in k_points:
    g0 = 1 \
        + np.exp(-1j * np.dot(k, d1)) \
        + np.exp(-1j * np.dot(k, d2)) \
        + np.exp(-1j * np.dot(k, d3))

    # 2x2 Hamiltonian
    H_k = np.array([
        [E_a, 4 * E_sa * g0],
        [4 * E_sa * np.conj(g0), E_c]
    ])

    eigvals = np.linalg.eigvalsh(H_k)
    E_minus.append(eigvals[0])
    E_plus.append(eigvals[1])
````

Our code define an **11×11 Hamiltonian**, so 1 extra orbital ( additional s* orbital) is included.


## 🧱 **2. The Tight-Binding Hamiltonian**

We use the **Slater-Koster tight-binding method**, which models:

- On-site energies $E$ for each orbital (like $E_{sa}, E_{pa}, \dots$)
    
- Hopping (off-diagonal) terms between orbitals, parameterized by **Slater-Koster integrals**:
    
    - $V_{ss\sigma}$
        
    - $V_{sp\sigma}$​
        
    - $V_{pp\sigma}, V_{pp\pi}$​, etc.
        

The hopping amplitude depends on:

- **Orbital types**
    
- **Relative direction between atoms**
    
- **Crystal symmetry**
    
- **Phase factors** $\exp(i \vec{k} \cdot \vec{R})$
    

These are encoded in the code through:

- - $g_0, g_1, g_2, g_3$: symmetry-weighted combinations of complex exponentials (structure factors)

## 📐 **3. FCC Crystal & Phase Factors**

FCC structure → each atom (As) is surrounded by **4 nearest-neighbor Ga atoms**, located at:

- $\vec{d}_1 = [0, 0, 0]$
    
- $\vec{d}_2 = [0, -1, -1]/2$
    
- $\vec{d}_3 = [-1, 0, -1]/2$
    
- $\vec{d}_4 = [-1, -1, 0]/2$
    

So we use:

$$p_{i}=e^{i\vec{k}\cdot \vec{d}_{i}}$$
$$g_{0}=\frac{1}{4}(p_{1}+p_{2}+p_{3}+p_{4}),~~(ss~\mathrm{and}~pp\sigma~\mathrm{hopping})$$
$$g_{1},g_{2},g_{3}=\mathrm{antisymmetric~combinations~(directional~hopping)}$$
## 🔄 **4. Band Formation**

Each orbital pair creates hybridization:

- **s-s** → via $V_{ss\sigma}$​
    
- **s-p** → via $V_{sp\sigma}$​
    
- **p-p** → via $V_{pp\sigma}, V_{pp\pi}$​
    

These matrix elements populate the **Hamiltonian matrix $H(\vec{k})$**.

You then diagonalize it for each $vec{k}$ to get **energy eigenvalues = band energies**.

---

## 📊 **5. $\vec{k}$-Direction:**

This model chooses a **line in reciprocal space** (L-direction: $[1, 1, 1]$):

- You vary $\vec{k} = k_{\text{max}} \cdot \vec{r} \cdot \frac{Nk}{Nt - 1}​$
    
- This gives a 1D slice through 3D $\vec{k}$-space.
    
#### **The General Idea**

 
 
 You’re generating **a sequence of $\vec{k}$-points** along that direction:
$$\vec{k}_{n}=n\cdot\Delta k\cdot \hat{r}$$
Where: 
- $n = 0, 1, 2, ..., Nt-1$
    
- $\hat{r} = \frac{[1, 1, 1]}{\sqrt{3}}$​ is the **direction unit vector**
    
- $\Delta k$ is the step size in magnitude
    
- The range is typically up to $k_{\text{max}}=\pi$
    

  
  
  But since you’re not normalizing the direction vector here (which is okay), you simply scale the full vector directly.



In order to ensure proper scaling, we normalize: $$\vec{k}_{n}=\frac{n}{Nt-1}\cdot \vec{r}_{unit}\cdot k_{max}$$

## 🛠️ **6. Bandgap Correction**

- $E[4]$: top of valence band
    
- $E[5]$: bottom of conduction band
    

Since empirical tight-binding might not hit the exact gap, this shifts **conduction bands only** upward.

````python
import numpy as np
import matplotlib.pyplot as plt

# Constants (in eV)
Esa = -8.3431
Epa = 1.0414
Esc = -2.6569
Epc = 3.6686
Esea = 8.5914
Esec = 6.7386

Vss = -6.4513
Vxx = 1.9546
Vxy = 5.0779
Vsapc = 4.4800
Vpasc = 5.7839
Vseapc = 4.8422
Vpasec = 4.8077

# k-direction settings
l, m, n = 1, 1, 1   # L-direction
kmax = np.pi
Nt = 21             # Number of k-points

# FCC basis vectors (choose only one set, both are equivalent)
d1 = np.array([0.0, 0.0, 0.0]) / 2
d2 = np.array([0.0, -1.0, -1.0]) / 2
d3 = np.array([-1.0, 0.0, -1.0]) / 2
d4 = np.array([-1.0, -1.0, 0.0]) / 2

# Arrays to store results
E = np.zeros((Nt, 11))
X = np.zeros(Nt)

for Nk in range(Nt):
    k = np.array([l, m, n]) * kmax * Nk / (Nt - 1)
    
    # Phase factors
    p1 = np.exp(1j * np.dot(k, d1))
    p2 = np.exp(1j * np.dot(k, d2))
    p3 = np.exp(1j * np.dot(k, d3))
    p4 = np.exp(1j * np.dot(k, d4))

    g0 = (p1 + p2 + p3 + p4) / 4
    g1 = (p1 + p2 - p3 - p4) / 4
    g2 = (p1 - p2 + p3 - p4) / 4
    g3 = (p1 - p2 - p3 + p4) / 4

    # Construct Hamiltonian (11x11)
    h = np.zeros((11, 11), dtype=complex)

    # Fill non-zero elements of the Hamiltonian
    h[0, 0] = Esa / 2
    h[0, 1] = Vss * g0
    h[0, 2] = Vsapc * g1
    h[0, 3] = Vsapc * g2
    h[0, 4] = Vsapc * g3

    h[1, 1] = Esc / 2
    h[1, 2] = -Vpasc * np.conj(g1)
    h[1, 3] = -Vpasc * np.conj(g2)
    h[1, 4] = -Vpasc * np.conj(g3)

    h[2, 2] = Epa / 2
    h[2, 5] = Vxx * g0
    h[2, 6] = Vxy * g3

    h[3, 3] = Epa / 2
    h[3, 5] = Vxy * g2
    h[3, 7] = Vxx * g0

    h[4, 4] = Epa / 2
    h[4, 6] = Vxy * g1
    h[4, 7] = Vxy * g2

    h[5, 5] = Epc / 2
    h[5, 8] = Vseapc * g1

    h[6, 6] = Epc / 2
    h[6, 8] = Vseapc * g2

    h[7, 7] = Epc / 2
    h[7, 8] = Vseapc * g3

    h[8, 8] = Esea / 2
    h[9, 9] = Esec / 2
    h[10, 10] = Esec / 2

    # Make Hermitian
    H = h + h.conj().T - np.diag(h.diagonal())

    # Diagonalize
    eigvals, _ = np.linalg.eigh(H)
    E[Nk, :] = np.sort(np.real(eigvals))

    # k-axis value
    X[Nk] = -(Nk) / (Nt - 1)  # L-direction

# Plotting
plt.figure(figsize=(10, 6))
for i in range(11):
    plt.plot(X, E[:, i], 'b', linewidth=2)
     # Plot mirrored data
    plt.plot(-X, E[:, i], 'b', linewidth=2, )  # alpha for lighter color

plt.xlabel("k (as fraction of maximum value) →", fontsize=14)
plt.ylabel("Energy (eV) →", fontsize=14)
plt.title("GaAs Band Structure (L-direction)", fontsize=16)
plt.xlim(-1,1)
plt.grid(True)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
````

````python
import numpy as np
import plotly.graph_objects as go

# Constants (same as your original code) — shortened here
Esa = -8.3431
Epa = 1.0414
Esc = -2.6569
Epc = 3.6686
Esea = 8.5914
Esec = 6.7386
Vss = -6.4513
Vxx = 1.9546
Vxy = 5.0779
Vsapc = 4.4800
Vpasc = 5.7839
Vseapc = 4.8422
Vpasec = 4.8077

# FCC basis vectors
d1 = np.array([0.0, 0.0, 0.0]) / 2
d2 = np.array([0.0, -1.0, -1.0]) / 2
d3 = np.array([-1.0, 0.0, -1.0]) / 2
d4 = np.array([-1.0, -1.0, 0.0]) / 2
ds = [d1, d2, d3, d4]

# K-grid
Nk = 50
kx_vals = np.linspace(-np.pi, np.pi, Nk)
ky_vals = np.linspace(-np.pi, np.pi, Nk)
KX, KY = np.meshgrid(kx_vals, ky_vals)
energy_band = np.zeros((Nk, Nk))

for i in range(Nk):
    for j in range(Nk):
        kx, ky = KX[i, j], KY[i, j]
        k = np.array([kx, ky, 0])  # 2D slice with kz = 0

        # Phase factors
        p = [np.exp(1j * np.dot(k, d)) for d in ds]
        g0 = sum(p) / 4
        g1 = (p[0] + p[1] - p[2] - p[3]) / 4
        g2 = (p[0] - p[1] + p[2] - p[3]) / 4
        g3 = (p[0] - p[1] - p[2] + p[3]) / 4

        # Build Hamiltonian
        h = np.zeros((11, 11), dtype=complex)
        h[0, 0] = Esa / 2
        h[0, 1] = Vss * g0
        h[0, 2] = Vsapc * g1
        h[0, 3] = Vsapc * g2
        h[0, 4] = Vsapc * g3

        h[1, 1] = Esc / 2
        h[1, 2] = -Vpasc * np.conj(g1)
        h[1, 3] = -Vpasc * np.conj(g2)
        h[1, 4] = -Vpasc * np.conj(g3)

        h[2, 2] = Epa / 2
        h[2, 5] = Vxx * g0
        h[2, 6] = Vxy * g3

        h[3, 3] = Epa / 2
        h[3, 5] = Vxy * g2
        h[3, 7] = Vxx * g0

        h[4, 4] = Epa / 2
        h[4, 6] = Vxy * g1
        h[4, 7] = Vxy * g2

        h[5, 5] = Epc / 2
        h[5, 8] = Vseapc * g1

        h[6, 6] = Epc / 2
        h[6, 8] = Vseapc * g2

        h[7, 7] = Epc / 2
        h[7, 8] = Vseapc * g3

        h[8, 8] = Esea / 2
        h[9, 9] = Esec / 2
        h[10, 10] = Esec / 2

        H = h + h.conj().T - np.diag(h.diagonal())
        eigvals, _ = np.linalg.eigh(H)
        energy_band[i, j] = np.real(np.sort(eigvals)[9])  # pick the 5th band (approx valence top)

# Plotly 3D surface
fig = go.Figure(data=[go.Surface(z=energy_band, x=KX, y=KY, colorscale='Viridis')])
fig.update_layout(title="Valence Band Surface (GaAs, kz=0)", scene=dict(
    xaxis_title="kx",
    yaxis_title="ky",
    zaxis_title="Energy (eV)"
))
fig.show()
````

````python
import plotly.graph_objects as go
import numpy as np

# FCC lattice parameter (arbitrary units)
a = 1

# Lattice vectors for FCC
lattice_vectors = np.array([
    [0, 0, 0],
    [0, 0.5, 0.5],
    [0.5, 0, 0.5],
    [0.5, 0.5, 0],
])

# Ga atoms at FCC positions
ga_atoms = lattice_vectors * a

# As atoms shifted by (¼,¼,¼)
as_atoms = (lattice_vectors + [0.25, 0.25, 0.25]) * a

# Plot
fig = go.Figure()

# Ga atoms
fig.add_trace(go.Scatter3d(
    x=ga_atoms[:, 0], y=ga_atoms[:, 1], z=ga_atoms[:, 2],
    mode='markers',
    marker=dict(size=6, color='blue'),
    name='Ga'
))

# As atoms
fig.add_trace(go.Scatter3d(
    x=as_atoms[:, 0], y=as_atoms[:, 1], z=as_atoms[:, 2],
    mode='markers',
    marker=dict(size=6, color='red'),
    name='As'
))

# Set layout
fig.update_layout(
    title='Zinc Blende Unit Cell (GaAs)',
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='z',
        aspectmode='cube'
    ),
    legend=dict(x=0.8, y=0.9)
)

fig.show()
````
